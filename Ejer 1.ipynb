```python
from EyR import *
```

```python
def ecuacionPendulo(t, ci, g, l, b):
    dOdt = ci[1]
    dvdt = (-(g/l)*numpy.sin(ci[0]))-(b*ci[1])
    return numpy.array([dOdt, dvdt])
```

```python
g = 9.8
l = 18.7
b = 0.2
ci = numpy.array([3, 0])
solucion, tiempo = euler(0, 80, ci, 0.01, ecuacionPendulo, g, l, b)
```

```python
pl.plot(tiempo, solucion[:,0], label="tiempo-angulo")
# pl.plot(tiempo, solucion[:,1], label="Poblacion de Chitas")
pl.xlabel("Tiempo")
pl.ylabel("Angulo")
pl.title("Tiempo-Angulo")
pl.grid()
pl.legend()
pl.show()
```

```python
pl.plot(tiempo, solucion[:,1], label="tiempo-velocidad")
# pl.plot(tiempo, solucion[:,1], label="Poblacion de Chitas")
pl.xlabel("Tiempo")
pl.ylabel("Velocidad")
pl.title("Tiempo-Velocidad")
pl.grid()
pl.legend()
pl.show()
```

```python
pl.plot(solucion[:,0], solucion[:,1], label="angulo-velocidad")
# pl.plot(tiempo, solucion[:,1], label="Poblacion de Chitas")
pl.xlabel("Angulo")
pl.ylabel("Velocidad")
pl.title("Angulo-Velocidad")
pl.grid()
pl.legend()
pl.show()
```

Segun las graficas se puede apreciar que el pendulo se detiene aprox en el segundo 70
